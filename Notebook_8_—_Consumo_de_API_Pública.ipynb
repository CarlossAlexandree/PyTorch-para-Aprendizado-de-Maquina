```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulação dos dados de câmbio
dados_simulados = {
    'BRL': 5.32,
    'EUR': 0.92,
    'JPY': 158.31,
    'BTC': 0.0000145,
    'GBP': 0.79
}

df = pd.DataFrame(dados_simulados.items(), columns=['Moeda', 'Cotação'])

# Visualização
plt.bar(df['Moeda'], df['Cotação'], color='dodgerblue')
plt.title("Cotação simulada de moedas em relação ao dólar")
plt.ylabel("Valor")
plt.show()
```
