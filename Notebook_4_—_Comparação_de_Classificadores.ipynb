```python
# Regressão Logística vs Árvore de Decisão vs SVM
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
df = pd.DataFrame({
    'gold_diff': [1200, 3000, -500, 2500, -1000, 1800],
    'kills_diff': [5, 8, -2, 6, -3, 4],
    'dragons_red': [1, 0, 3, 1, 2, 1],
    'winner': [1, 1, 0, 1, 0, 1]
})

X = df[['gold_diff', 'kills_diff', 'dragons_red']]
y = df['winner']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Modelos
models = {
    "Logistic": LogisticRegression(),
    "Árvore": DecisionTreeClassifier(),
    "SVM": SVC()
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc

# Visualização
plt.bar(results.keys(), results.values(), color=['blue', 'green', 'red'])
plt.title("Comparação de Acurácia")
plt.ylabel("Acurácia")
plt.show()
```
