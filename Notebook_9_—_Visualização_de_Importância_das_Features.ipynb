```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fictícios e importâncias simuladas
features = ['gold_diff', 'kills_diff', 'towers_diff', 'dragons_diff']
importancias = [0.52, 0.28, 0.15, 0.05]

df = pd.DataFrame({'Feature': features, 'Importância': importancias})
df = df.sort_values(by='Importância')

# Gráfico horizontal
plt.barh(df['Feature'], df['Importância'], color='teal')
plt.title("Importância das variáveis na predição")
plt.xlabel("Peso")
plt.ylabel("Feature")
plt.show()
```
