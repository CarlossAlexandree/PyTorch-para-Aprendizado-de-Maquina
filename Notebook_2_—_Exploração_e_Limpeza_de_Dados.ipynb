```python
# Limpeza de dataset de partidas
import pandas as pd

# Dataset simulado com inconsistências
df = pd.DataFrame({
    'gold_diff': [1200, None, 3000, -1000, '2500'],
    'team_winner': [1, 0, None, 0, 1],
    'kills_diff': ['5', '-2', 8, None, 3]
})

# Conversão de tipos
df['gold_diff'] = pd.to_numeric(df['gold_diff'], errors='coerce')
df['kills_diff'] = pd.to_numeric(df['kills_diff'], errors='coerce')

# Remoção de nulos
df = df.dropna()

# Estatísticas
print(df.describe())

# Visualização
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['gold_diff'], kde=True)
plt.title("Distribuição da Diferença de Ouro")
plt.show()
```
