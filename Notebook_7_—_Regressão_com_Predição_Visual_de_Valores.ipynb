```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Dataset fictício: relação entre tempo de jogo e ouro acumulado
df = pd.DataFrame({
    'tempo_min': [20, 25, 30, 35, 40, 45, 50],
    'ouro_total': [12000, 15000, 18000, 22000, 26000, 31000, 36000]
})

# Modelo de regressão
X = df[['tempo_min']]
y = df['ouro_total']
model = LinearRegression().fit(X, y)

# Predição
tempo_novo = np.array([[55]])
ouro_previsto = model.predict(tempo_novo)

# Visualização
plt.scatter(df['tempo_min'], df['ouro_total'], label='Dados reais')
plt.plot(df['tempo_min'], model.predict(X), color='red', label='Regressão')
plt.scatter(tempo_novo, ouro_previsto, color='green', label=f'Previsão: {int(ouro_previsto)}')
plt.title("Predição de ouro acumulado pelo tempo de partida")
plt.xlabel("Tempo (min)")
plt.ylabel("Ouro total")
plt.legend()
plt.show()
```
